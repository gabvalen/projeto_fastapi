```python
#calculando a precisão do nosso código (Acurácia)
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x = dados_evasao.drop(columns=['Travel_Time', 'Study_Time', 'Wants_Higher_Education', 'Internet_Access', 'In_Relationship', 'Going_Out', 'Weekend_Alcohol_Consumption', 'Weekday_Alcohol_Consumption', 'Grade_1', 'Grade_2', 'Dropped_Out'])
#retirei as colunas que não serão informadas na pesquisa e a coluna dropped out que é o resultado de nossa busca
y = dados_evasao['Dropped_Out']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) #80% para treinamento e 20% para teste

model = DecisionTreeClassifier()

# Apply one-hot encoding to the training and testing data
x_train_encoded = pd.get_dummies(x_train)
x_test_encoded = pd.get_dummies(x_test)

# To ensure both train and test sets have the same columns after encoding,
# reindex the test set to match the training set columns.
missing_cols_test = set(x_train_encoded.columns) - set(x_test_encoded.columns)
for c in missing_cols_test:
    x_test_encoded[c] = 0
x_test_encoded = x_test_encoded[x_train_encoded.columns]


# adjust the model for training
model.fit(x_train_encoded, y_train)
predictions = model.predict(x_test_encoded)
score = accuracy_score(y_test, predictions)
score
#nosso modelo está com acurácia de 100%. Uma das razões é que a base de dados é pequena e a exclusão de algumas colunas de dados.
```
